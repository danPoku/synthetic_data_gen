```python
from toolkit.core import SyntheticDataGenerator
from toolkit.utils import save_to_csv, preview_data, plot_monthly_trends
from toolkit.diseases import load_disease_definitions
import pandas as pd
```

```python
# Initialize and generate synthetic data
generator = SyntheticDataGenerator("Accra", "2015-01-01", "2024-12-31")
df = generator.simulate()
```

```python
df = pd.DataFrame(df)
```

```python
# Export generated data to csv
save_to_csv(df, "respiratory_cases_accra.csv")
```

```python
disease_list = [
    "Upper Respiratory Tract Infection (J00–J06)",
    "Pneumonia (J12–J18)",
    "Bronchiolitis (J21)",
    "Acute Bronchitis (J20)",
    "Asthma (J45)",
    "Influenza (J09–J11)",
    "Chronic Obstructive Pulmonary Disease (J44)"
]
```

```python
# Preview data for a selected disease
preview_data(df)

for disease in disease_list:
    plot_monthly_trends(df, disease)
```
